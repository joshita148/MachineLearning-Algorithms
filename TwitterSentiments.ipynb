```python
import string
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
stop_words = set(stopwords.words('english'))
stop_words.update([i for i in string.punctuation])
```

```python
stop_words.update(["'s"])
```

```python
stop_words
```

```python
from nltk.corpus import wordnet
def simple_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADJ
    else:
        return wordnet.NOUN
```

```python
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag
lemmatizer= WordNetLemmatizer()
```

```python
a= "abx1"
a.isalpha()
```

```python
def clean(words):
    clean_words=[]
    for w in words:
        if w.lower() not in stop_words and w.isalpha():
            pos= pos_tag([w])
            clean_word= lemmatizer.lemmatize(w, pos= simple_pos(pos[0][1]))
            clean_words.append(clean_word)
    return " ".join(clean_words)
```

```python
train_text_document=[]
for text in text_train:
    clean_text= clean(word_tokenize(text))
    train_text_document.append(clean_text)
```

```python
test_text_document=[]
for text in text_test:
    clean_text= clean(word_tokenize(text))
    test_text_document.append(clean_text)
```
